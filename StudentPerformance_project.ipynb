**Import Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**Load Dataset**

```python
df = pd.read_csv("StudentsPerformance.csv")

print("Dataset Loaded")
print(df.head())
```

**Dataset Info**

```python
print(df.info())
```

**Check Null Values**

```python
print(df.isnull().sum())
```

**Basic Statistics**

```python
print(df.describe())
```

**Math Score Distribution Graph**

```python
plt.figure()
sns.histplot(df["math score"], bins=30)
plt.title("Math Score Distribution")
plt.show()
```

**Reading Score Distribution**

```python
plt.figure()
sns.histplot(df["math score"], bins=30)
plt.title("Math Score Distribution")
plt.show()
```

**Writing Score Distribution**

```python
plt.figure()
sns.histplot(df["writing score"], bins=30)
plt.title("Writing Score Distribution")
plt.show()
```

**Gender count Graph**

```python
plt.figure()
sns.countplot(x="gender", data=df)
plt.title("Gender Count")
plt.show()
```

**Lunch Type Graph**

```python
plt.figure()
sns.countplot(x="lunch", data=df)
plt.title("Lunch Type Count")
plt.show()
```

**Test Preparation Course Graph**

```python
plt.figure()
sns.countplot(x="test preparation course", data=df)
plt.title("Test Preparation Count")
plt.show()
```

**Math vs Reading Scatter plot**

```python
plt.figure()
sns.scatterplot(x="math score", y="reading score", data=df)
plt.title("Math vs Reading")
plt.show()
```

**Math vs Writing Scatter plot**

```python
plt.figure()
sns.scatterplot(x="math score", y="writing score", data=df)
plt.title("Math vs Writing")
plt.show()
```

**Correlation Heatmap**

```python
plt.figure()
sns.heatmap(df.corr(numeric_only=True), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```

**Boxplot Math Score**

```python
plt.figure()
sns.boxplot(x=df["math score"])
plt.title("Math Score Boxplot")
plt.show()
```

**Boxplot Reading Score**

```python
plt.figure()
sns.boxplot(x=df["reading score"])
plt.title("Reading Score Boxplot")
plt.show()
```

**Boxplot Writing Score**

```python
plt.figure()
sns.boxplot(x=df["writing score"])
plt.title("Writing Score Boxplot")
plt.show()
```

**Average Score Column**

```python
df["average"] = (df["math score"] + df["reading score"] + df["writing score"]) / 3

print(df.head())
```

**Average Score Histogram**

```python
plt.figure()
sns.histplot(df["average"])
plt.title("Average Score Distribution")
plt.show()
```

**Pass fail Columns**

```python
df["result"] = df["average"].apply(lambda x: "Pass" if x >= 40 else "Fail")

print(df.head())
```

**Pass Fail Graph**

```python
plt.figure()
sns.countplot(x="result", data=df)
plt.title("Pass vs Fail")
plt.show()
```
